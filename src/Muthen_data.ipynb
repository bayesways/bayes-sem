```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
url = 'https://www.statmodel.com/examples/baysem/BHPS%20OINDRESPX1.DAT'
df0 = pd.read_table(url, sep='\t', delim_whitespace=True,  header=None)
df0
```

```python
colnames = ['y'+str(x) for x in range(1,16)]
colnames.extend(['sex', 'age'])
```

```python
df = df0.iloc[:,pd.np.r_[0:15, 23, 27]]
df.columns = colnames
df.head()
```

```python
df_age_filter = df[(df.age +47.4 <= 55.057) &(df.age +47.4 >=50)]
df_age_filter.sex.value_counts()
```

```python
df_final = df_age_filter.iloc[:,:-1].astype(int)
df_final
```

```python
df_final_men = df_final[df.sex == 1].iloc[:,:-1]
df_final_men.to_csv("../dat/muthen_men.csv", index=False)
```

```python
df_final_women = df_final[df.sex == 2].iloc[:,:-1]
df_final_women.to_csv("../dat/muthen_women.csv", index=False)
```
