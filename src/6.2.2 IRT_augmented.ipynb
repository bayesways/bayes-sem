```python
import numpy as np
import pandas as pd
import pystan
from scipy.stats import norm, multivariate_normal, invwishart, invgamma, bernoulli
from scipy.special import expit

import datetime
import sys
import os

from codebase.plot import * 
from codebase.file_utils import save_obj, load_obj

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
np.random.seed(121)
nsim_data = 200
J = 6
K = 2
beta = np.array([[1,0], [-2, 2],[-1,-1],[2,1], [3,-1], [1,-1]], dtype=float)
mu = np.array([1,2,.3,-.8, 1, -1.4])
Sigma = np.eye(J)*0.04
uu = multivariate_normal.rvs(mean = np.zeros(J), cov = Sigma, size=nsim_data)
zz = multivariate_normal.rvs(mean = np.zeros(K), cov=np.eye(K), size=nsim_data)
yy = mu + zz@beta.T
DD = bernoulli.rvs(p=expit(yy)); DD

# err = np.abs((beta@beta.T + Sigma) - np.cov(yy.T))
```

```python
data = dict()
data['N'] = nsim_data
data['K'] = K
data['J'] = J
data['Sigma'] = Sigma
data['y'] = yy
data['D'] = DD
data['beta'] = beta
data['u'] = uu
data['mu'] = mu
data['Omega'] = data['beta']@data['beta'].T 
```

```python
stan_data = dict(N = data['N'], K = data['K'], J = data['J'], DD = data['D'])
```

```python
# confirm that the credible intervals cover 0 for all elements of u
def get_non_zeros(x, prc_min = 10, prc_max = 90):
    """
    Returns the index of the elements that do not contain
    zero in their quantile interval from prc_min to prc_max.
    Indices are returned in two arrays a1, a2. The i-th non
    zero element of the matrix x is at position [a1[i], a2[i]].
    """
    rcs = np.percentile(x, [prc_min,prc_max], axis=0)
    min_b = rcs[0,:,:]
    max_b = rcs[1,:,:]

    zeros = np.zeros((x.shape[1], x.shape[2]))

    indx = (min_b < zeros) & (zeros < max_b)
    # data['u'][~indx]
    return np.nonzero(~indx)

get_non_zeros(ps[0])
```
