```python
def gen_data(nsim_data, rho =0.5, random_seed=None):
    if random_seed is not None:
        np.random.seed(random_seed)

    alpha = np.array([2, -.8,])
    sigma = np.array([1.2, .7])
    Phi_corr = np.eye(2)
#     Phi_corr[0,1] = rho
#     Phi_corr[1,0] = rho
    Phi = np.diag(sigma) @ Phi_corr @  np.diag(sigma)

    yy = multivariate_normal.rvs(mean = alpha, cov=Phi, size=nsim_data)

    data = dict()
    data['N'] = nsim_data
    data['K'] = 2
    data['alpha'] = alpha
    data['sigma'] = sigma
    data['Phi_corr'] = Phi_corr
    data['Phi'] = Phi
    data['y'] = yy
    data['sigma_prior'] = np.diag(np.linalg.inv(np.cov(data['y'], rowvar=False)))


    return(data)

data = gen_data(100)
data
```

```python
sm = pystan.StanModel(model_code=model_code)
```

```python
stan_data = dict(N = data['N'], K = data['K'], yy = data['y'], sigma_prior = data['sigma_prior'])
```

```python
fit = sm.sampling(data=stan_data, iter=3000, chains=1)
```

```python
print(fit)
```

```python
ps= fit.extract(permuted=False, pars=['alpha', 'sigma'])  # return a dictionary of arrays
ps['alpha']
```

```python
ps['alpha'] = np.squeeze(ps['alpha'])
ps['sigma'] = np.squeeze(ps['sigma'])
```
